```python
import subprocess
from pathlib import Path

# =========================
# 설정
# =========================
input_file = Path("나는 SOLO.E243.260304.720p-NEXT.mp4")
output_file = Path("나는 SOLO.E243.260304.720p-NEXT.mp4")

start_time = "00:40:00"   # 시작 시간
end_time = "00:49:44"     # 끝 시간

# =========================
# ffmpeg로 영상 자르기
# =========================
cmd = [
    "ffmpeg",
    "-y",                  # 기존 파일 덮어쓰기
    "-i", str(input_file), # 입력 파일
    "-ss", start_time,     # 시작 시간
    "-to", end_time,       # 끝 시간
    "-c:v", "libx264",     # 영상 재인코딩
    "-c:a", "aac",         # 오디오 재인코딩
    str(output_file)
]

print("영상 자르기 시작...")
result = subprocess.run(cmd, capture_output=True, text=True)

if result.returncode == 0:
    print(f"완료: {output_file}")
else:
    print("오류 발생")
    print(result.stderr)
```
