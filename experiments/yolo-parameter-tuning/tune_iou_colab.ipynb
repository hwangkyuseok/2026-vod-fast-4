## 1. iou 튜닝 실험 실행

```
import warnings
warnings.filterwarnings('ignore')

import time
import numpy as np
import pandas as pd
from pathlib import Path
from ultralytics import YOLO
from tqdm.auto import tqdm

# =========================
# 기본 설정
# =========================
MODEL_PATH = "yolov8l.pt"
VIDEO_PATH = DRIVE_SCENE_PATH
DATA_YAML = "coco128.yaml"

# MVP 15 class ids
MVP_CLASS_IDS = [16, 26, 39, 41, 45, 53, 56, 57, 59, 60, 62, 63, 65, 67, 72]

# 고정 파라미터
FIXED_IMGSZ = 640

# conf 후보 (conf 튜닝 결과에서 선정)
CONF_CANDIDATES = [0.25, 0.40]

# iou 실험 범위: 0.3 ~ 0.7 (0.1 간격)
IOU_VALUES = [0.3, 0.4, 0.5, 0.6, 0.7]

total_experiments = len(CONF_CANDIDATES) * len(IOU_VALUES)

# =========================
# 모델 로드
# =========================
print("Loading YOLOv8L model...")
model = YOLO(MODEL_PATH)

scene_files = sorted(Path(VIDEO_PATH).glob("*.mp4"))
print(f"Scene clips: {len(scene_files)}개")
print(f"conf 후보: {CONF_CANDIDATES}")
print(f"iou 실험값: {IOU_VALUES}")
print(f"총 실험 횟수: {total_experiments}회 ({len(CONF_CANDIDATES)} conf x {len(IOU_VALUES)} iou)")
print("=" * 55)
```

```
rows = []
exp_num = 0

for conf_val in CONF_CANDIDATES:
    for iou_val in IOU_VALUES:
        exp_num += 1
        exp_id = f"I{exp_num}"

        print(f"\n{'='*55}")
        print(f"  [{exp_num}/{total_experiments}] conf={conf_val}, iou={iou_val}, imgsz={FIXED_IMGSZ}")
        print(f"{'='*55}")

        # --- COCO128 Validation ---
        print("  [1/2] COCO128 Validation...")
        val_start = time.time()
        metrics = model.val(
            data=DATA_YAML,
            imgsz=FIXED_IMGSZ,
            conf=conf_val,
            iou=iou_val,
            classes=MVP_CLASS_IDS,
            verbose=False
        )
        val_time = time.time() - val_start

        precision_arr = metrics.box.p
        recall_arr = metrics.box.r
        precision = float(np.nanmean(precision_arr)) if hasattr(precision_arr, "__len__") else float(precision_arr)
        recall = float(np.nanmean(recall_arr)) if hasattr(recall_arr, "__len__") else float(recall_arr)
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

        print(f"  Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f} ({val_time:.1f}s)")

        # --- SampleVideo_Scenes 추론 (stream=True) ---
        print(f"  [2/2] SampleVideo_Scenes 추론 ({len(scene_files)} clips)...")
        total_detected = 0
        start_time = time.time()

        pbar = tqdm(scene_files, desc=f"  conf={conf_val} iou={iou_val}", leave=True)
        for scene_file in pbar:
            results = model.predict(
                source=str(scene_file),
                conf=conf_val,
                iou=iou_val,
                imgsz=FIXED_IMGSZ,
                classes=MVP_CLASS_IDS,
                stream=True,
                verbose=False
            )
            for result in results:
                if result.boxes is not None:
                    total_detected += len(result.boxes)
            pbar.set_postfix({"detected": total_detected})

        elapsed = time.time() - start_time
        det_per_sec = total_detected / elapsed if elapsed > 0 else 0

        print(f"  => 탐지: {total_detected}개 | 시간: {elapsed:.1f}s | {det_per_sec:.1f} det/s")

        rows.append({
            "실험": exp_id,
            "conf": conf_val,
            "iou": iou_val,
            "imgsz": FIXED_IMGSZ,
            "Precision": round(precision, 4),
            "Recall": round(recall, 4),
            "F1": round(f1, 4),
            "탐지 수": total_detected,
            "소요 시간(s)": round(elapsed, 1),
            "탐지/sec": round(det_per_sec, 1)
        })

df = pd.DataFrame(rows)
print("\n" + "=" * 55)
print("  모든 실험 완료!")
print("=" * 55)
```

## 2. 결과 확인

```
from IPython.display import display

# F1 최고값 강조
best_idx = df["F1"].idxmax()
best_row = df.loc[best_idx]

def highlight_best(row):
    if row.name == best_idx:
        return ['background-color: #90EE90; font-weight: bold'] * len(row)
    return [''] * len(row)

styled = df.style.apply(highlight_best, axis=1).format({
    'conf': '{:.2f}',
    'iou': '{:.1f}',
    'Precision': '{:.4f}',
    'Recall': '{:.4f}',
    'F1': '{:.4f}',
    '탐지/sec': '{:.1f}'
})

print("=== conf=0.25 그룹 ===")
display(df[df['conf'] == 0.25].style.format({
    'Precision': '{:.4f}', 'Recall': '{:.4f}', 'F1': '{:.4f}'
}))

print("\n=== conf=0.40 그룹 ===")
display(df[df['conf'] == 0.40].style.format({
    'Precision': '{:.4f}', 'Recall': '{:.4f}', 'F1': '{:.4f}'
}))

print(f"\n★ Best F1 = {best_row['F1']} @ conf={best_row['conf']}, iou={best_row['iou']}")
print(f"  Precision={best_row['Precision']}, Recall={best_row['Recall']}")
print(f"  탐지 수={best_row['탐지 수']}, 소요 시간={best_row['소요 시간(s)']}s")
```

```
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

colors = {0.25: 'royalblue', 0.40: 'tomato'}
markers = {0.25: 'o', 0.40: 's'}

# --- 왼쪽: F1 비교 ---
for conf_val in CONF_CANDIDATES:
    subset = df[df['conf'] == conf_val]
    axes[0].plot(subset['iou'], subset['F1'], f'-{markers[conf_val]}',
                color=colors[conf_val], label=f'conf={conf_val}', linewidth=2, markersize=8)

axes[0].axvline(x=best_row['iou'], color='green', linestyle='--', alpha=0.4)
axes[0].set_xlabel('IoU Threshold', fontsize=12)
axes[0].set_ylabel('F1 Score', fontsize=12)
axes[0].set_title('F1 Score vs IoU', fontsize=13)
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# --- 가운데: Precision vs Recall ---
for conf_val in CONF_CANDIDATES:
    subset = df[df['conf'] == conf_val]
    axes[1].plot(subset['iou'], subset['Precision'], f'--{markers[conf_val]}',
                color=colors[conf_val], label=f'P (conf={conf_val})', linewidth=1.5, alpha=0.7)
    axes[1].plot(subset['iou'], subset['Recall'], f'-{markers[conf_val]}',
                color=colors[conf_val], label=f'R (conf={conf_val})', linewidth=2)

axes[1].set_xlabel('IoU Threshold', fontsize=12)
axes[1].set_ylabel('Score', fontsize=12)
axes[1].set_title('Precision & Recall vs IoU', fontsize=13)
axes[1].legend(fontsize=9)
axes[1].grid(True, alpha=0.3)

# --- 오른쪽: 탐지 수 ---
x = np.arange(len(IOU_VALUES))
width = 0.35
for i, conf_val in enumerate(CONF_CANDIDATES):
    subset = df[df['conf'] == conf_val]
    axes[2].bar(x + i*width, subset['탐지 수'].values, width,
               color=colors[conf_val], alpha=0.8, label=f'conf={conf_val}')

axes[2].set_xlabel('IoU Threshold', fontsize=12)
axes[2].set_ylabel('Total Detections', fontsize=12)
axes[2].set_title('Detection Count vs IoU', fontsize=13)
axes[2].set_xticks(x + width/2)
axes[2].set_xticklabels([str(v) for v in IOU_VALUES])
axes[2].legend(fontsize=11)
axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('iou_tuning_result.png', dpi=150, bbox_inches='tight')
plt.show()
print("그래프 저장: iou_tuning_result.png")
```
